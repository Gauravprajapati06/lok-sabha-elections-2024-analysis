```python
table_data=data.find("table")
rows=table_data.find_all('tr',class_='tr')    ##finding the data that belongs to element tr, within table element
```

```python
party_names=[]
for row in rows:
    curr_party_name=[]
    for cell in row.find_all("td"):
        if cell.string!="\n" and cell.string!=None:
            curr_party_name.append(cell.string)
        else:
            curr_party_name.append("NA")
    party_names.append(curr_party_name)
print(party_names)                #this code scrapes the name of the party and the total seats won by each party.
```

```python
df=pd.DataFrame()
for i in party_names:
    df=df.append({'Party name': i[0],
                  'seats won': i[3]},ignore_index=True)
df.head()                        ##converts the above data to data frames
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_election_data(site_url):
    # Send an HTTP GET request to retrieve the HTML content
    http_response = requests.get(site_url)

    # Check if the request was successful
    if http_response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        html_soup = BeautifulSoup(http_response.content, 'html.parser')

        # Find all tables with the specified class
        all_tables = html_soup.find_all('table', class_='table table-striped table-bordered')

        # Proceed if tables are found
        if all_tables:
            # Extract table headers, excluding those with colspan
            header_list = [th.get_text(strip=True) for th in all_tables[0].find_all('th') if not th.has_attr('colspan')]

            # Initialize a list to store table rows
            table_rows = []

            # Iterate over each table to extract row content
            for table in all_tables:
                for table_row in table.find('tbody').find_all('tr'):
                    row_data = []
                    for table_cell in table_row.find_all('td', recursive=False):
                        # Check for nested tables within cells
                        inner_table = table_cell.find('table')
                        if inner_table:
                            inner_cell = inner_table.find('td')
                            row_data.append(inner_cell.get_text(strip=True) if inner_cell else "")
                        else:
                            row_data.append(table_cell.get_text(strip=True))
                    table_rows.append(row_data)

            # Filter rows to ensure they match the header count
            rows_cleaned = [row for row in table_rows if len(row) == len(header_list)]

            # Create and return a DataFrame
            return pd.DataFrame(rows_cleaned, columns=header_list)
        else:
            print("No tables found.")
            return pd.DataFrame()
    else:
        print(f"Error retrieving HTML: Status code {http_response.status_code}")
        return pd.DataFrame()
```

# AndhraPradesh

```python
res_1_ap=fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS011.htm")
res_2_ap=fetch_election_data("https://results.eci.gov.in/PcResultGenJune2024/statewiseS012.htm")
data_1_ap= pd.DataFrame(res_1_ap)                    ##converting the fetched data to dataframes
data_2_ap=pd.DataFrame(res_2_ap)
final_ap=pd.concat([data_1_ap, data_2_ap])
```

```python
final_ap.to_csv('andhrapradesh.csv', sep=',', index=False, encoding='utf-8') #converting the required dataframes to csv files for visualizations
```
